```python
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
from scipy.misc import derivative
import math
import seaborn
seaborn.set(style='ticks')

# sigmoid graph
x = np.arange(-8, 8, 0.1)
f = 1 / (1 + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
# tanh graph
x = np.arange(-8, 8, 0.1)
f = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, num=100)

fx = []
for i in range(len(x)):
    fx.append((x[i]-1)**2 * (x[i]+1))
    
plt.plot(x, fx)
plt.grid()
plt.axvline()
plt.axhline()
plt.show()
```

```python
# taylor approximation example
def plot():
    x_lims = [-5,5]
    x1 = np.linspace(x_lims[0],x_lims[1],800)

    fx = []
    for i in range(len(x1)):
#         fx.append(-x1[i]**2+1)
        fx.append((x1[i]-2)**2 * (x1[i]+0.1))
#     plt.plot(x1,np.cos(x1),label='f(x)')
    gx = []
    for i in range(len(x1)):
        gx.append(-(1.5*x1[i]-1)**2 + 1.5)
        
    plt.plot(x1,fx, label='f(x)', marker='.', markevery=[1,1])
    plt.plot(x1, gx, label='g2(x)')
    plt.xlim(x_lims)
    plt.ylim([-5,5])
    plt.axvline(linewidth=0.5, color='k')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

plot()
```

```python
# example of linear approximation
def f(x):
    return np.exp(x)
#     return x**2

# gradient
def d(x):
    return derivative(f,x)

x1 = 1
y1 = f(x1)

# tangent line
def line(x, x1, y1):
    return d(x1)*(x - x1) + y1

# range for the tangent 
xrange = np.linspace(x1-1, x1+5, 10)
```

```python
# range for the function
y = np.linspace(-3,3,num=100)

plt.figure()
plt.plot(y, f(y))
plt.axvline(linewidth=0.5, color='k')
plt.axhline(linewidth=0.5, color='k')
plt.scatter(x1, y1, color='C1', s=50)
plt.xticks([])
plt.yticks([])
plt.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2)
```

```python
# another example of linear approximation
def fig_sixteen(x):
    return x**2

def fig_sixteen_grad(x):
    return derivative(fig_sixteen, x)

def tangent_fig(x, x1, y1):
    return fig_sixteen_grad(3)*(x - 3) + 9
x_11 = 3
y_11 = fig_sixteen(x_11)

y = np.linspace(-5,6,100)

plt.figure()
plt.plot(y, fig_sixteen(y))
plt.axvline(linewidth=0.5, color='k')
plt.axhline(linewidth=0.5, color='k')
plt.scatter(x_11, y_11, color='C1', s=50)
plt.plot(xrange, tangent_fig(xrange, x_11, y_11), 'C1--', linewidth = 2)
```

```python
# example for multivariate linear approximation 
def f(x, y):
    return (x ** 2 + y ** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 500)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
# maybe useful figure to talk about maximum points
def f(x, y):
    return np.sin(x) * np.sin(y) * np.exp(-x**2 - y**2)

x = np.linspace(-6, 6, 500)
y = np.linspace(-6, 6, 500)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.Spectral,
                       linewidth=0, antialiased=False)
ax.view_init(30, 45)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# plt.show()
plt.savefig('filename.png', dpi=300)

```
